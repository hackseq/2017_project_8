```python
import pandas as pd
import numpy as np
```

```python
data = np.load("../data/hackseq_features.npy")
column_names = open("../data/feature_names.txt").readlines()
```

```python
trimmed_column_names = [col.strip() for col in column_names][:-1]
```

```python
df = pd.DataFrame(data=data[:,:-1], columns=trimmed_column_names)
print(df.shape)
```

```python
melted_df = pd.melt(frame = df, 
        id_vars = [col for col in list(df.columns) if col not in ['AZD_200nM', 'PLX_2uM', '6TG_2ug/mL']],
        var_name = "drug",
        value_name = "lfc")
```

```python
melted_df.to_csv("../results/melted_hackseq.csv")
```
